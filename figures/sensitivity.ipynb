# Sensitity Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f1_scores = np.array([0.757, 0.8, 0.805, 0.865, 0.849, 0.88, 0.849])
budgets = ["64", "128", "192", "256", "384", "512", "1024"]

fig, ax = plt.subplots(figsize=(8,4))

ax.bar(budgets, f1_scores, width=0.5)
ax.grid(which='major', axis='y', linestyle='--')
ax.set_xticks(budgets)
ax.set_xlabel('Budget [-]', fontsize=14)
ax.set_ylim([0.7, 1])
ax.set_ylabel(r'$F_1$-Score [-]', fontsize=14)

#fig.savefig('sensitivity_f1.pdf')
```

```python
normal = np.array([28, 49, 81, 131, 222, 343, 852])
normal_b = np.array([32,64,96,128,192,256,0])
normal = normal + normal_b
anomaly = np.array([36, 79, 111, 125, 162, 169, 172])
sum = normal + anomaly
normal = normal/sum * 100
anomaly = anomaly/sum * 100

fig, ax = plt.subplots(figsize=(6,4))

ax.bar(budgets, anomaly, label='Anomalien', width=0.5)
ax.bar(budgets, normal, bottom=anomaly, label='Normalfälle', width=0.5)

ax.set_xticks(budgets)
ax.grid(which='major', axis='y', linestyle='--')
ax.set_xlabel('Budget [-]', fontsize=14)
ax.set_ylabel('Anteil [%]', fontsize=14)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)

#fig.savefig("sensitivity_proportions.pdf")
```

```python
f1_scores = np.array([0.757, 0.8, 0.805, 0.865, 0.849, 0.88, 0.849])
budgets = ["64", "128", "192", "256", "384", "512", "1024"]

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,4))

rect = ax2.bar(budgets, f1_scores, width=0.5)
ax2.grid(which='major', axis='y', linestyle='--')
ax2.set_xticks(budgets)
ax2.set_xlabel('Budget [-]', fontsize=14)
ax2.set_ylim([0.7, 1])
ax2.set_ylabel(r'$F_1$-Score [-]', fontsize=14)

ax2.bar_label(rect, padding=3)

ax2.text(0.5, 1.15, "(b)", transform=ax2.transAxes, horizontalalignment='center', size=16)
ax1.text(0.5, 1.15, "(a)", transform=ax1.transAxes, horizontalalignment='center', size=16)

normal = np.array([28, 49, 81, 131, 222, 343, 852])
normal_b = np.array([32,64,96,128,192,256,0])
normal = normal + normal_b
anomaly = np.array([36, 79, 111, 125, 162, 169, 172])
sum = normal + anomaly
normal = normal/sum * 100
anomaly = anomaly/sum * 100

ax1.bar(budgets, anomaly, label='Anomalien', width=0.5)
ax1.bar(budgets, normal, bottom=anomaly, label='Normalfälle', width=0.5)
ax1.set_xticks(budgets)
ax1.grid(which='major', axis='y', linestyle='--')
ax1.set_xlabel('Budget [-]', fontsize=14)
ax1.set_ylabel('Anteil [%]', fontsize=14)
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)

fig.savefig('sensitivity.pdf', bbox_inches='tight')
```
