```python
f1_unsupervised = np.array([0.545, 0.286])
f1_supervised = np.array([0.865, 0.849])
datasets = ["Energiedaten", "mechanische\nProzessdaten"]
x = np.arange(len(datasets))

fig, ax = plt.subplots(figsize=(6,5))

width=0.4
rect2 = ax.bar(x-width/2, f1_unsupervised, width=width, label='un√ºberwachtes\nModell')
rect1 = ax.bar(x+width/2, f1_supervised, width=width, label='aktives Lernen')

ax.grid(which='major', axis='y', linestyle='--')
ax.set_xticks(x)
ax.set_xticklabels(datasets, fontsize=14)
ax.set_ylim([0.2, 1])
ax.set_ylabel(r'$F_1$-Score [-]', fontsize=16)
ax.legend(ncol=2, loc='lower center', bbox_to_anchor=(0.5,1), fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=14)

ax.bar_label(rect1, padding=3, fontsize=14)
ax.bar_label(rect2, padding=3, fontsize=14)

fig.tight_layout()

fig.savefig('energy-mechanical_large.pdf')
```
