# Unsupervised Training

```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
with open('../saved_models/cnn-lstm/history.json') as f:
    history = json.load(f)

train_loss = np.log(history['loss'])
val_loss = np.log(history['val_loss'])
train_loss = history['loss']
val_loss = history['val_loss']

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(train_loss[3:], label=r'$MSE_{train}$')
ax.plot(val_loss[3:], label=r'$MSE_{valid}$')

ax.set_xlabel('Epochen [-]', fontsize=14)
ax.set_ylabel(r'$MSE$ [-]', fontsize=14)

ax.grid(which='major', axis='y', linestyle='--')
ax.legend(fontsize=14)
fig.tight_layout()

fig.savefig('unsupervised_training.pdf')
```
