# Benefit of active Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f1_unsupervised = np.array([0.444, 0.545, 0.557, 0.500, 0.500])
f1_supervised = np.array([0.771, 0.849, 0.816, 0.800, 0.833])
archs = ["LSTM", "CNN", "CNN-LSTM", "CNN-LSTM\n(Self-Attention)", "WaveNet"]
x = np.arange(len(archs))

fig, ax = plt.subplots(figsize=(8,4))

width=0.3
rect1 = ax.bar(x-width/2, f1_unsupervised, width=width, label='un√ºberwachtes\nModell')
rect2 = ax.bar(x+width/2, f1_supervised, width=width, label='aktives Lernen')
ax.grid(which='major', axis='y', linestyle='--')
ax.set_xticks(x)
ax.set_xticklabels(archs)
ax.set_ylim([0.2, 1])
ax.set_ylabel(r'$F_1$-Score [-]', fontsize=14)
ax.legend(loc='upper left')

ax.bar_label(rect1, padding=3)
ax.bar_label(rect2, padding=3)

fig.tight_layout()

fig.savefig('active_improvement.pdf', bbox_inches='tight')
```
