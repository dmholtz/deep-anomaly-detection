# Optimization of the latent space dimension

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
latent_dims = np.arange(start=1, stop=17)
validation_loss = np.array([0.0037,	0.0026,	0.0037,	0.0028,	0.0018,	0.0014,	0.0013,	0.0013,	0.0013,	0.0012,	0.0014,	0.0013,	0.0010,	0.0011,	0.0013,	0.0013])

fig, ax = plt.subplots(figsize=(10,4))

ax.plot(latent_dims, validation_loss, 'x:')

ax.grid(which='major', linestyle='--')
ax.vlines(6, 0, 0.005, linestyles='--', colors='tab:orange', label=r'$d_{z, opt}$')
ax.set_xticks(latent_dims)
ax.set_ylim([0.0008, 0.004])
ax.set_xlabel(r'Dimension des latenten Raums $d_z$ [-]', fontsize=14)
ax.set_ylabel(r'$MSE_{valid}$ [-]', fontsize=14)
ax.legend(fontsize=14)
fig.tight_layout()

fig.savefig('latent_dim_opt.pdf')
```
