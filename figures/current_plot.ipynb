```python
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get the working directory of the ipykernel
working_dir = os.getcwd()
working_path = Path(working_dir)

# define subdirectories
repo_path = working_path.parent.absolute()
train_data_path = os.path.join(repo_path, 'data', 'energy', 'train')

x = np.load(os.path.join(train_data_path, "x.npy"))
```

```python
sample = 0
feature = 1

lower = 4.2
upper = 12.3

current = x[sample,:375,feature]
time = np.linspace(0, 18, num=current.shape[0])
time_ticks = np.arange(0, 20, step=2)

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(time, current, label=r'$i_{robot}$')
ax.set_xticks(time_ticks)
ax.set_xlabel("Zeit [s]", fontsize=14)
ax.set_ylabel("Gesamtstrom [A]", fontsize=14)
ax.set_ylim([0.15, 0.7])

ax.axvline(lower, linestyle='--', c='k')
ax.axvline(upper, linestyle='--', c='k')

ax.text(lower/2/18,0.9, "Anfahrt", fontsize=14, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
ax.text((lower+upper)/2/18,0.9, "Montage", fontsize=14, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
ax.text((upper+(18-upper)/2)/18,0.9, "RÃ¼ckfahrt", fontsize=14, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)

fig.tight_layout()
fig.savefig('current_plot.pdf', bbox_inches='tight')
```
