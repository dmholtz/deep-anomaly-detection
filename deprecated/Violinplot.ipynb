```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

n = 5
x1 = np.random.normal(size=n)
x2 = np.random.normal(loc=5, size=n)

df1 = pd.DataFrame({'score': x1, 'info': ['one']*n})
df2 = pd.DataFrame({'score': x2, 'info': ['two']*n})
df = pd.concat([df1, df2])
print(df)

df['info'] = df['info'].astype('category')

ax = sns.violinplot(data=df, y='score', hue='info', split=True)

#ax = sns.violinplot(x="info", y="score", hue="info",
#                    data=df, palette="Set2", split=True, scale="count")

df['info']

#plt.hist(df)
```
