```python
def mahalanobis_anomaly_score(x_true, x_pred):
    
    error = np.abs(x_true-x_pred)
    print(error.shape)
    
    mu = np.mean(error, axis=0)
    cov = np.cov(error)
    
    a = np.zeros((error.shape[0],))
    for i in range(len(a)):
        e = error[i]
        #a[i] = (e-mu).T*np.linalg.inv(cov)*(e-mu)
        a[i] = (e-mu).T*1/cov*(e-mu)
    return a

x_true = np.sin(np.linspace(0,2*np.pi,100))+ 0.05*np.sin(10*np.linspace(0,6,100))
x_pred = np.sin(np.linspace(0,2*np.pi,100))

import matplotlib.pyplot as plt

plt.plot(x_true)
plt.plot(x_pred)
a = mahalanobis_anomaly_score(x_true, x_pred)
plt.plot(a)
plt.show()
```
